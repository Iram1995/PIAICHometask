5. Calculate Volume of a sphere
Write a Python program to get the volume of a sphere, please take the radius as input from user

```python
import math
radius = float(input("Enter radius: "))
volume = ((4/3)*math.pi)*(radius*radius*radius)
print("Volume of Sphere with radius", radius ,"is",round(volume,2))
```

```python
def decimalToBinary(n): 
    return bin(n).replace("0b","") 
decimal = int(input("Enter a decimal number: "))
binary = decimalToBinary(decimal)
print("Binary Representation of ",decimal," is ",binary)
```

```python
def binaryToDecimal(binary): 
    return int(binary,2) 
binary = input("Enter a binary number: ")
decimal = binaryToDecimal(binary)
print("Decimal Representation of ",binary," is ",decimal)
```

```python
text = input("Enter text: ")
vowels = ['a','A','e','E','i','I','o','O','u','U']
vowelsCount = 0
consonantCount = 0
listString = [str(x) for x in str(text)] 
for i in listString:
    if i in vowels:
        vowelsCount = vowelsCount + 1
    else:
        consonantCount = consonantCount+1

print("Vowels: " , vowelsCount)
print("Consonants: " , consonantCount)
```

```python
text = str(input("Enter text: "))
reversedText =  text[::-1]
if(text == reversedText):
    print("Text ",text," is Palindrome")
else:
     print("Text ",text," is not Palindrome")
    
```
